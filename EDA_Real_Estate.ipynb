```python
# merapihkan data
md_MissVal = md.dropna()
md_MissVal.isna().sum()
```

```python
# Melihat distribusi data kolom Price dan Landsize
md_MissVal.hist(figsize=(14,14), xrot=0)
plt.show()
```

```python
# 
md_MissVal_describe = md_MissVal[['Price','Landsize']].describe()
md_MissVal_describe
```

```python
# Menghitung Central Tendency menggunakan median(), 
# karena lebih cocok digunakan untuk distribusi miring
md_MissVal_CenTend = md_MissVal[['Price','Landsize']].median()
md_MissVal_CenTend
```

```python
# Menggunakan perhitungan standar deviasi
md_MissVal_dispersion = md_MissVal[['Price','Landsize']].std()
md_MissVal_dispersion
```
